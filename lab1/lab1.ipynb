```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()
```

```python
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
#11
iris_target_0 = iris[iris['target'] == 0]
iris_target_1 = iris[iris['target'] == 1]
iris_target_2 = iris[iris['target'] == 2]
label = ["setosa",'versicolor', 'virginica']
fig = plt.figure()
fig.set_size_inches(10,10)
bins = np.linspace(0.8,7.2,50)
plt.subplot(2,2,1)
plt.hist([iris_target_0['petal length (cm)'],iris_target_1['petal length (cm)'],iris_target_2['petal length (cm)']], bins, rwidth=10, label = label)
plt.legend()
plt.title("petal length (cm)")
plt.subplot(2,2,2)
bins = np.linspace(0,2.6,26)
plt.hist([iris_target_0['petal width (cm)'],iris_target_1['petal width (cm)'],iris_target_2['petal width (cm)']], bins, label = label)
plt.legend()
plt.title("petal width (cm)")
plt.subplot(2,2,3)
bins = np.linspace(4.4,7.5,30)
plt.hist([iris_target_0['sepal length (cm)'],iris_target_1['sepal length (cm)'],iris_target_2['sepal length (cm)']], bins, label = label)
plt.legend()
plt.title("sepal length (cm)")
plt.subplot(2,2,4)
bins = np.linspace(2,4.4,25)
plt.hist([iris_target_0['sepal width (cm)'],iris_target_1['sepal width (cm)'],iris_target_2['sepal width (cm)']], bins, label = label)
plt.legend()
plt.title("sepal width (cm)")
plt.show()
```
